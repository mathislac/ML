```python
import glob
import os

# merging the files
files_joined = os.path.join('../batchs', "restaurant*.csv")

# Return a list of all joined files
list_files = glob.glob(files_joined)

print("** Merging multiple csv files into a single pandas dataframe **")
# Merge files by joining all files
df2 = pd.concat(map(pd.read_csv, list_files), ignore_index=True)
df2.head()
```

```python
df2["ca"] = df2["Quantity"] * df2["Product Price"]
df2.head()
```

```python
df3 = df2[['Order Date','Order Number', 'ca']]
df3.head()
```

```python
df4 = df3.groupby(['Order Number','Order Date'])['ca'].sum()
df4.head()
```

```python
df4['Order Date'] = pd.to_datetime(df3['Order Date'])
df4.head()
```
